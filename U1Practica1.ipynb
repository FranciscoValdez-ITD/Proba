```python
from statistics import *
```

```python
# Lista de calificaciones del grupo a de probabilidad y estadistica
grupoa=[70,70,65,90,100,95,45,80,73,20,100,96]
print(grupoa)
type(grupoa)
```

```python
# Sacar la media de las calificaciones con una Formula
promedio=sum(grupoa)/len(grupoa)
print(promedio)
```

```python
# Determinar el valor de la mediana con la funcion median
mediana=median(grupoa)
print(mediana)
```

```python
# Determinar el valor de la moda con la funcion mode
moda=mode(grupoa)
print(moda)
```

```python
# En caso que exista mas de una moda, se puede emplear la funcion multimode
# Multiples modas que pueden existir
moda2=multimode(grupoa)
print(moda2)
```

```python
# Definimos una variable llamada tiempo100mts de tipo lista, que contiene los
#tiempos que tardaron los 10 alumnos para correr
tiempo100mts=[15.10,17.20,14.69,13.27,22.15,18.71,19.15,20.65,15.10,17.20]

# Desplegamos los valores de la lista
print(tiempo100mts)
```

```python
# Usamos la funcion mean para determinar el tiempo promedio de la lista de valores
print("El tiempo promedio de los corredores es:", mean(tiempo100mts))
print("La media de los corredores es:", median(tiempo100mts))
print("La moda de los corredores es:", mode(tiempo100mts))
```

```python
# Libreria para generar numeros aleatorios
import random
```

```python
fabrica1=[]
for i in range(100):
    numero_random=round(random.uniform(50,100),2)
    fabrica1.append(numero_random)
print (fabrica1)
```

```python
fabrica2=[]
for i in range(100):
    numero_random2=round(random.uniform(70,100),2)
    fabrica2.append(numero_random2)
print (fabrica2)
```

```python
# Libreria de Python para graficar valores
import matplotlib.pyplot as plt
```

```python
plt.hist(fabrica1)
plt.title("Produccion diaria de Fabrica 1")
```

```python
from typing import Text
plt.hist(fabrica2)
plt.title("Produccion diaria de Fabrica 2")
```

```python
media1=mean(fabrica1)
mediana1=median(fabrica1)
moda1=mode(fabrica1)

print("Valores de la Fabrica 1")
print("Media:",media1)
print("Mediana:",mediana1)
print("Moda:",moda1)
```

```python
media2=mean(fabrica2)
mediana2=median(fabrica2)
moda2=mode(fabrica2)

print("Valores de la Fabrica 2")
print("Media:",media2)
print("Mediana:",mediana2)
print("Moda:",moda2)
```

```python
# Generar 2 subplot de 1 renglon por 2 columnas para desplegar los 2 histogramas
# una al lado de la otra
fig, (ax1,ax2)=plt.subplots(nrows=1,ncols=2,figsize=(10,5))
ax1.hist(fabrica1)
ax1.set_title("Produccion diaria de Fabrica 1")
ax2.hist(fabrica2)
ax2.set_title("Produccion diaria de Fabrica 2")
plt.show()
```
