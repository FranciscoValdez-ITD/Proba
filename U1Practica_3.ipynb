```python
# Basado en el libro de probabilidad y estadistica
import random
import statistics as stats
edades=[]
for i in range(100):
 na=random.randint(18,70)
 edades.append(na)
print(edades)
```

```python
import pandas as pd
```

```python
# Crear un Dataframe de pandas con los numeros aleatorios
df=pd.DataFrame(edades,columns=['Edad'])
df.head()
```

```python
# Calcular la tabla de distribucion de frecuencias
tabla_frecuencias = df['Edad'].value_counts().sort_index()
# Convertir la Serie a un DataFrame y resetear el indice para que 'Edad' sea una columna
tabla_frecuencias = tabla_frecuencias.reset_index()
# Renombrar las columnas
tabla_frecuencias.columns = ['Edad', 'Frecuencia']

# Imprimir la tabla
print("Tabla de distribucion de frecuencias")
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Relativa']=tabla_frecuencias['Frecuencia']/100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Porcentual']=tabla_frecuencias['Frecuencia']/100*100
tabla_frecuencias
```
