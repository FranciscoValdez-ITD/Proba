```python
# Basado en el libro de probabilidad y estadistica
import random
import statistics as stats
edades=[]
for i in range(100):
 na=random.randint(18,70)
 edades.append(na)
print(edades)
```

```python
import pandas as pd
```

```python
# Crear un Dataframe de pandas con los numeros aleatorios
df=pd.DataFrame(edades,columns=['Edad'])
df.head()
```

```python
# Calcular la tabla de distribucion de frecuencias
tabla_frecuencias = df['Edad'].value_counts().sort_index()
# Convertir la Serie a un DataFrame y resetear el indice para que 'Edad' sea una columna
tabla_frecuencias = tabla_frecuencias.reset_index()
# Renombrar las columnas
tabla_frecuencias.columns = ['Edad', 'Frecuencia']

# Imprimir la tabla
print("Tabla de distribucion de frecuencias")
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Relativa']=tabla_frecuencias['Frecuencia']/100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Porcentual']=tabla_frecuencias['Frecuencia']/100*100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Acumulada']=tabla_frecuencias['Frecuencia']/100*100
tabla_frecuencias
```

```python
# Importamos la libreria math de funciones matematicas
import math
```

```python
# Volvemos a unir la lista de edades
print(edades)
```

```python
df = pd.DataFrame({'Edad': edades})
df
```

```python
n=len(edades)
k=int(1+math.log2(n))
print("Valor de n= ",n,"\numero de intervalos0",k)
```

```python
# Crear tabla de frecuencias
tabla_frecuencias=pd.cut(df['Edad'],bins=k,include_lowest=True).value_counts().reset_index()
tabla_frecuencias.columns=['Clase','Frec']
tabla_frecuencias
```

```python
# Ordenar Tabla
tabla_frecuencias=tabla_frecuencias.sort_values(by=['Clase'])
tabla_frecuencias
```

```python
# Generar columnas siguientes
tabla_frecuencias['Frec Relativa']=tabla_frecuencias['Frec'] /n
tabla_frecuencias['Frec Porcentual']=tabla_frecuencias['Frec'] /n*100
tabla_frecuencias['Frec Acumulada']=tabla_frecuencias['Frec'].cumsum()
tabla_frecuencias['Frec Rel Ac']=tabla_frecuencias['Frec Relativa'].cumsum()
tabla_frecuencias['Frec Porc Ac']=tabla_frecuencias['Frec Porcentual'].cumsum()
# Mostrar la tabla
print("Tabla de distribucion de frecuencias")
tabla_frecuencias
```

```python
import matplotlib.pyplot as plt
plt.hist(df['Edad'],bins=k,alpha=0.5)
plt.title('Histograma de edades')
plt.xlabel('Edades')
plt.ylabel('Frecuencia')
```
