```python
# Importamos las librerías necesarias
import random
import matplotlib.pyplot as plt
from statistics import *
```

```python
# Generamos datos aleatorios para la fábrica 1
fabrica1 = []

for i in range(100):
    numero_aleatorio = round(random.uniform(50, 100), 2)
    fabrica1.append(numero_aleatorio)

print(fabrica1)
```

```python
# Generamos datos aleatorios para la fábrica 2
fabrica2 = []

for i in range(100):
    numero_aleatorio = round(random.uniform(40, 93), 2)
    fabrica2.append(numero_aleatorio)

print(fabrica2)
```

```python
# Diagrama de caja de la fábrica 1
plt.boxplot(fabrica1)
plt.ylabel('Producción en mts.')
plt.title('Producción de la fábrica 1')
plt.show()
```

```python
# Diagramas de caja comparativos de ambas fábricas
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))

ax1.boxplot(fabrica1, vert=True, patch_artist=True)
ax1.set_title('Producción fábrica 1')

ax2.boxplot(fabrica2, vert=True, patch_artist=True)
ax2.set_title('Producción fábrica 2')

plt.show()
```

```python
# Insertamos un valor atípico en la fábrica 1
fabrica1.append(150)
print(fabrica1)
```

```python
# Diagramas de caja con valor atípico incluido
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))

ax1.boxplot(fabrica1, vert=True, patch_artist=True)
ax1.set_title('Producción fábrica 1')

ax2.boxplot(fabrica2, vert=True, patch_artist=True)
ax2.set_title('Producción fábrica 2')

plt.show()
```

```python
# Diagramas de caja con escala vertical uniforme
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9, 5))

ax1.boxplot(fabrica1, vert=True, patch_artist=True)
ax1.set_title('Producción fábrica 1')
ax1.set_ylim(30, 160)

ax2.boxplot(fabrica2, vert=True, patch_artist=True)
ax2.set_title('Producción fábrica 2')
ax2.set_ylim(30, 160)

plt.show()
```
