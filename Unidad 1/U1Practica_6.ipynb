## Estadística descriptiva del Dataset Autos

En esta práctica se realiza un análisis estadístico descriptivo del dataset
de automóviles, utilizando medidas de tendencia central, histogramas y
diagramas de caja para analizar el comportamiento del precio de los autos.

```python
# Importar libreria
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
autos = pd.read_csv("/autos2.csv")
autos.head()
```

```python
autos = autos.dropna(subset=["price"])
autos["price"] = pd.to_numeric(autos["price"])
autos["price"].min()
autos["price"].max()
autos["price"].mode()
autos["price"].mean()
autos["price"].median()
autos["price"].std()
autos["price"].describe()
```

```python
n = autos["price"].count()
bins = int(1 + 3.322 * np.log10(n))
bins
```

```python
plt.hist(autos["price"], bins=bins)
plt.title("Histograma de Precios de Automóviles")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.show()
```

```python
sns.boxplot(x="body-style", y="price", data=autos)
plt.title("Precio vs Tipo de Vehículo")
plt.xlabel("Tipo de Vehículo")
plt.ylabel("Precio")
plt.show()
```

```python
sns.boxplot(x="drive-wheels", y="price", data=autos)
plt.title("Precio vs Tipo de Tracción")
plt.xlabel("Tracción")
plt.ylabel("Precio")
plt.show()
```

```python
sns.boxplot(x="engine-location", y="price", data=autos)
plt.title("Precio vs Ubicación del Motor")
plt.xlabel("Ubicación del Motor")
plt.ylabel("Precio")
plt.show()
```

## 6. Conclusiones generales

- El precio de los automóviles presenta una distribución asimétrica.
- Existen valores atípicos en los precios más altos.
- El tipo de vehículo influye significativamente en el precio.
- Los autos con tracción trasera y motor trasero tienden a tener precios más elevados.
- Los diagramas de caja facilitan la comparación visual entre categorías.
